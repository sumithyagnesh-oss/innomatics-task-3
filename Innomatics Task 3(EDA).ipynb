```python
#Maximum Product of Two Elements in an Array
class Solution:
    def maxProduct(self, nums):
        result = 0
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                result = max(result, (nums[i] - 1) * (nums[j] - 1))
        return result
```

```python
s1=Solution()
a=[3,6,23,7,4,7]
print(s1.maxProduct(a))
```

```python
#Count Number of Teams
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        count = 0
        
        for j in range(n):
            left_less = left_greater = right_less = right_greater = 0
            
            for i in range(j):
                if rating[i] < rating[j]:
                    left_less += 1
                elif rating[i] > rating[j]:
                    left_greater += 1
            
            for k in range(j + 1, n):
                if rating[k] < rating[j]:
                    right_less += 1
                elif rating[k] > rating[j]:
                    right_greater += 1
            
            count += left_less * right_greater + left_greater * right_less
        
        return count
```

```python
s2=Solution()
b=[3,6,2,7,4]
print(s2.numTeams(b))
```

```python
#Number of Students Doing Homework at a Given Time
class Solution:
    def busyStudent(self, startTime,endTime,queryTime):
        result=0
        for i in range(len(startTime)):
            if startTime[i]<=queryTime<=endTime[i]:
                result+=1
        return result
```

```python
s3=Solution()
x=[3,6,2,3,4]
y=[5,7,3,6,7]
print(s3.busyStudent(x,y,6))
```

```python
#Number of Steps to Reduce a Number to Zero
class Solution:
    def numberOfSteps(self, num):
        step=0
        while(num!=0):
            if num % 2==0:
                num/=2
            else:
                num-=1
            step+=1
        return step
```

```python
s4=Solution()
print(s4.numberOfSteps(19))
```

```python
#Counting Bits
class Solution:
    def countBits(self, n):
        result = [0] * (n + 1)
        for i in range(1, n + 1):
            result[i] = result[i >> 1] + (i & 1)
        return result
```

```python
s5=Solution()
print(s5.countBits(19))
```
